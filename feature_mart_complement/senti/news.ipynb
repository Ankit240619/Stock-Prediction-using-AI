```python
import os
import requests
from tqdm import tqdm
import zipfile

# 下载 GDELT 数据的函数
def download_gdelt_data(start_year, end_year, output_dir):
    base_url = "http://data.gdeltproject.org/events/"
    file_list_url = "http://data.gdeltproject.org/events/index.html"
    
    # 创建存储目录
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    # 循环下载每年的数据
    for year in range(start_year, end_year + 1):
        for month in range(1, 13):
            for day in range(1, 32):
                date_str = f"{year}{month:02d}{day:02d}"
                file_url = f"{base_url}{date_str}.export.CSV.zip"
                local_path = os.path.join(output_dir, f"{date_str}.export.CSV.zip")
                
                # 跳过已下载的文件
                if os.path.exists(local_path):
                    continue
                
                try:
                    response = requests.get(file_url, stream=True)
                    if response.status_code == 200:
                        with open(local_path, "wb") as f:
                            f.write(response.content)
                        print(f"Downloaded: {local_path}")
                except Exception as e:
                    print(f"Error downloading {file_url}: {e}")
```

```python
import pandas as pd
import glob

# 解压 ZIP 文件
def unzip_files(input_dir, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)
    
    for zip_file in glob.glob(f"{input_dir}/*.zip"):
        try:
            with zipfile.ZipFile(zip_file, 'r') as z:
                z.extractall(output_dir)
                print(f"Extracted: {zip_file}")
        except Exception as e:
            print(f"Error extracting {zip_file}: {e}")
```

```python
download_gdelt_data(start_year=2020, end_year=2024, output_dir="./gdelt_raw")
```
